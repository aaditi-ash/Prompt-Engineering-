```python
import PyPDF2

def extract_text_pypdf2(pdf_path):
    
    text = ""
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        for page in reader.pages:
            text += page.extract_text()
    return text


pdf_file = 'medicaldata.pdf'
data = extract_text_pypdf2(pdf_file)
print(data)
```
